# Part 8. POS Tagging
### Recall the notation: NN: Noun, singular or mass; NNS: Noun, plural; NNP: Proper noun, singular; NNPS: Proper noun, plural

```python
import nltk
```

```python
br = pd.read_csv(r'E:\MSBA\Courses\Spring\Text Analytics\Project\Dataset\Final\br_model_phx_2014-2018.csv')
```

```python
# Change string to list 
br['review_pos'] = br['review_words'].apply(lambda x: x[1:-1].split(','))
```

```python
br['review_pos'] = br['review_pos'].apply(lambda x: [item.replace("'", '') for item in x])
```

```python
# Remove space
br['review_pos'] = br['review_pos'].apply(lambda x: [item.replace(" ", '') for item in x])
```

```python
import string
def remove_punctuations(text):
    for punctuation in string.punctuation:
        if punctuation in text:
            text = text.replace(punctuation, '')
    return text
```

```python
# Remove punctuations
br['review_pos'] = br['review_pos'].apply(lambda x: [remove_punctuations(item) for item in x])
```

```python
# Remove empty string
br['review_pos'] = br['review_pos'].apply(lambda x: list(filter(None, x)))
```

```python
# Generate a list of POS tags with unprocessed tokens
br['review_pos'] = br['review_pos'].apply(lambda x: nltk.pos_tag(x))
```

```python
#Keep tags that start with 'N'
br['review_noun'] = br['review_pos'].apply(lambda x: [word for (word,tag) in x if tag.startswith('N')])
```

```python
br.to_csv(r'E:\MSBA\Courses\Spring\Text Analytics\Project\Dataset\Final\br_model_phx_2014-2018.csv', index=False)
```

# Part 11. Generate Term Document Matrix

```python
import gensim
from gensim import corpora,models
from gensim.models import LdaModel, LsiModel
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
br = pd.read_csv(r'E:\MSBA\Courses\Spring\Text Analytics\Project\Dataset\Final\br_model_phx_2014-2018.csv')
```

```python
br['review_model'] = br['review_model'].apply(lambda x: x[1:-1].split(','))
```

```python
# Remove space
br['review_model'] = br['review_model'].apply(lambda x: [item.replace(" ", '') for item in x])
```

```python
br['review_model'] = br['review_model'].apply(lambda x: [item.replace("'", '') for item in x])
```

```python
# Generate token dictionary class
dictionary = corpora.Dictionary(br['review_model']) 
print(dictionary)
```

```python
# Generate a unique token list 
sort_token = sorted(dictionary.items(),key=lambda k:k[0], reverse = False)
unique_token = [token for (ID,token) in sort_token]
```

```python
# Build a corpus
br['corpus'] = [dictionary.doc2bow(tokens) for tokens in br['review_model']]
print(br['corpus'].head())
# (id, tf in the row)
```

```python
# Save a Term Document Matrix
import numpy as np
matrix = gensim.matutils.corpus2dense(br['corpus'], num_terms=len(dictionary), dtype = 'int')
```

```python
matrix.T
#(all unique word's tf in the row)
```

```python
pd.DataFrame(matrix.T)
```

```python
# Transpose the matrix and convert the numpy matrix into pandas data frame
br2 = pd.concat([br,pd.DataFrame(matrix.T)], axis=1)
```

# Part 12. Topic modeling using LDA

```python
# Fit lda model
lda = models.LdaModel(br2['corpus'], id2word=dictionary, num_topics=10) 
# Topic matrix (V matrix)
lda.print_topics(10) 
```

```python
# Generate U Matrix for LDA model
corpus_lda = lda[br2['corpus']] #transform lda model

# Convert corpus_lda to numpy matrix
U_matrix_lda = gensim.matutils.corpus2dense(corpus_lda,num_terms=10).T
# (Topic weight in the row)

# Write U_matrix into pandas dataframe and output
br2 = pd.concat([br2,pd.DataFrame(U_matrix_lda)], axis=1)
```
